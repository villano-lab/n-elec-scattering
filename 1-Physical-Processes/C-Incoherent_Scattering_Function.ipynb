# Comparing Incoherent Scattering Functions
Scattering functions and structure effects are related to the cross section for neutron magnetic scattering. For an accurate cross section calculation, we need to determine if local electron behavior, random scattering etc. play a significant role in Neutron magnetic scattering probability. These are accounted for in incoherent scattering functions. 

 the electrons behave as individual scatterers. In the cross section calculations, we model electrons in the lattice or atomic structure assuming they behave as free electrons? <mark>Not sure if this is right ask Ant</mark> . Here, we present the argument for this assumption. 

Coherent scattering mostly abides by the atomic form factor, F(q,Z), and tends to be more sensitive to the overall lattice structure. Whereas the incoherent scattering function, S(q,Z), is more sensitive to individual electrons. In higher momentum transfer scattering events, such as relatively heavy neutrons interacting with electrons, the incoherent scattering function approaches the atomic number of the target, Z. So the range of momentum transfers we're interested in structure effects are negligible?

Below, incoherent scattering functions and experimental data for several target elements are plotted for comparison. The most relevant target elements are Silicon and Germanium, because they are common materials used as detectors for rare event searches.

## Hubbell and Freeman Comparison



```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import scipy.constants as sc

#print(sc.h)
# def x_to_theta(x):
#     return x*2/sc.hbar
def x_to_theta(x): #using equation 27 of Hubbell x=[sin]
    lambda_val=180e-12 #De Broglie wavelength of a neutron, from wikipedia
    theta = 2*np.arcsin(lambda_val*x)
    return theta


df_Si1['Theta']= x_to_theta(df_Si1['x'])
df_Si['Theta'] = x_to_theta(df_Si['X'])
df_Si3['Theta'] = x_to_theta(df_Si3['X'])
df_Si4['Theta'] = x_to_theta(df_Si4['X'])
df_Si_JB['Theta'] = x_to_theta(df_Si_JB['X'])
df_Ge['Theta'] = x_to_theta(df_Ge["X"])
df_Ge_WH['Theta'] = x_to_theta(df_Ge_WH['X'])
df_Ge1['Theta'] = x_to_theta(df_Ge1['x'])




plt.plot(df_Si['Theta'], df_Si['Y'], marker=r'$\theta$', linestyle='-', color='r', alpha=0.3, markersize=0.4, markeredgewidth=2, label ='Si')
plt.plot(df_Si3['Theta'], df_Si3['Y'], marker=r'$\theta$', linestyle='-', color='g', alpha=0.3, markersize=0.4, markeredgewidth=2, label = 'Si 3+')
plt.plot(df_Si4['Theta'], df_Si4['Y'], marker=r'$\theta$', linestyle='-', color='b', alpha=0.3, markersize=0.4, markeredgewidth=2, label ='Si 4+')
plt.plot(df_Si_JB['Theta'], df_Si_JB['Y'], marker=r'$\theta$', linestyle='-', color='m', alpha=0.3, markersize=0.4, markeredgewidth=2, label = 'Si JB')

plt.plot(df_Si1['Theta'], df_Si1['S(x,Z)'], label = "Si")
plt.plot(df_Ge1['Theta'], df_Ge1['S(X,Z)'], label = "Ge")

plt.plot(df_Ge['Theta'], df_Ge['Y'], marker='x', linestyle='-', color='r', alpha=0.3, markersize=0.4, markeredgewidth=2, label ='Ge')
plt.plot(df_Ge_WH['Theta'], df_Ge_WH['Y'], marker='x', linestyle='-', color='g', alpha=0.3, markersize=0.4, markeredgewidth=2, label = 'Ge Waller Hartree')

plt.xlabel('theta')
plt.ylabel('S(x,Z)')
plt.xscale('log')
plt.grid(True)
plt.legend()

print(df_Si)
```
